# KNN_IRIS_1D

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import neighbors, datasets
import pandas as pd
```

```python
path = '/content/drive/MyDrive/AI/AIO_Advanced/TEMP/2. KNN/code - update/iris_1D.csv'

data = pd.read_csv(path)
data
```

```python
x = data['Petal_Length'].to_numpy()
x = x.reshape(6,1)
y = data['Label'].to_numpy()
print(x,'\n',y)

x_test = [[2.4]]
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(x,y)
y_pred = classifier.predict(x_test)
print(y_pred)
```

# KNN_Iris 2D

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
path = '/content/drive/MyDrive/AI/AIO_Advanced/TEMP/2. KNN/code - update/iris_2D.csv'
#/content/drive/MyDrive/AI/AIO_Advanced/TEMP/2. KNN/code - update/iris_2D_mm.csv

data = pd.read_csv(path)
data
```

```python
X = data[['Petal_Length', 'Petal_Width']].to_numpy().reshape(6,2)
y = data['Label'].to_numpy()
print(X,'\n',y)

X_test = [[2.6, 0.7]]
```

```python
classifier = KNeighborsClassifier(n_neighbors=1)
classifier.fit(X,y)
y_pred = classifier.predict(X_test)
print(y_pred)
```
